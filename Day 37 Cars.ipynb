```python
# Create list with top 10 Categories

top10=[x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
print(top10)
```

```python
for label in top10:
    print(label)
```

```python
# Do one hot encoding to column X2 - only for top 10 Categories

for label in top10:
    data[label]=np.where(data['X2']==label,1,0)
print(data[['X2']+top10].head(20))    
```

```python
# Repeat the same with multiple columns with for Loop

cprajwal=pd.read_csv('mercedesbenz.csv',usecols=['X1','X2','X3','X4','X5','X6'])
for col in cprajwal.columns:
    top10=[x for x in cprajwal[col].value_counts().sort_values(ascending=False).head(10).index]
    print(col)
    print(top10)
    for label in top10:
        cprajwal[label]=np.where(cprajwal[col]==label,1,0)
    print(cprajwal[[col]+top10].head(20))    
    
```
